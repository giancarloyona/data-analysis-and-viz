## Part 2

Please recreate the bar chart seen below.  

* The figure should be 8x6
* The two colors used are `#74b9ff` and `#7ed6df`



```python
# Use this data to generate the plot!
resorts = ["Crested Butte", "Aspen", "Telluride", "Wolf Creek"]
snow_last_24 = [29,15,15,23]
snow_next_24 = [16,6,19,30]
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,6), dpi=200)
plt.title("CO Snow Storm Report")
plt.bar(resorts, snow_last_24, color="#74b9ff", label="Last 24hrs")
plt.bar(resorts, snow_next_24, color="#7ed6df", bottom=snow_last_24, label="Next 24hrs")
plt.ylabel("Inches")
plt.legend()
plt.show()
```

Recreate the pie chart seen below:

* I've provided lists for the actual numbers, the labels, and the colors
* Figure should be 8x8
* Notice the percentages in each slice!
* Two slices are slightly exploded



```python
# Use these lists to generate the chart!
labels = [
    "Building Construction", 
    "Deferred Maintenance",
    "General Contractor Fees",
    "Owner Costs",
    "Site Construction"
]
costs = [
    47945975,
    3753584,
    7340612,
    18247137,
    5766685
]
colors = ["#355070", "#6d597a", "#b56576", "#e56b6f", "#eaac8b"]
```

```python
figure = plt.figure(figsize=(8,8), dpi=200)
plt.pie(costs, labels=labels, autopct="%2.f%%", colors=colors, explode=(0,0.1,0,0,0.1))
plt.title("School District Cost Breakdown")
plt.show()
```

## Part 4

Recreate the figure seen in this image:

* Use the `kc_house_data.csv` dataset (import it with Pandas)
* The figure is 10x4


```python
# Do your work here!
import pandas as pd
```

```python
df = pd.read_csv('../datasets/kc_house_data.csv', parse_dates=['date'])
```

```python
sales_by_weekday = df['date'].dt.weekday.value_counts().sort_index()
days_week = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

months = df['date'].dt.month.unique()
months.sort()

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.bar(days_week, sales_by_weekday)
plt.title("Sales By Week Day")

plt.subplot(1,2,2)
plt.plot(months, df['date'].dt.month.value_counts().sort_index())
plt.title("Sales By Month")
plt.xticks(months)
plt.xlabel("Month")
plt.ylabel("Home Sales")
plt.show()
```
